```python
#Sort using a particular column
A = A.sort_values('Location')
A.head(10)
```

```python
#Sort the data using 2 columns
A = A.sort_values(['Location','Price'])
A.head()
```

```python
#Subset data where number of bedrooms = 3
B = A[A['Bedrooms']==3]
B.head(5)
```

```python
#Subset data where Location is Oceano
C = A[A['Location']=='Oceano']
C
```

```python
#Aggregate the data to understand what is the average House Size in each location
D = A.groupby(['Location'])[['Size']].mean()
D.head()
```

```python
#Aggregate the data to understand the number of houses in each location
E = A.groupby(['Location'])[['MLS']].count()
E.head()
```

```python
#Merge the above two data sets (B and C) together using inner join based on “Location” column
F = pd.merge(D,E,on='Location')
F.head()
```

```python
#Write the last file (F) into your directory
F.to_csv('F.csv')
```
